```python
class Library:
    
    def __init__(self,booksList,name_of_library,lended_books = []):
        
        self.booksList = booksList # all books in the library (available and not available).
        self.name_of_library = name_of_library
        self.lended_books = lended_books # list of tuples, the tuple (name of user, book).
        self.__available_books = [] # the books which are NOT lended.
        self.__not_available_books = [] # the books which are lended. (only without the name of user)
        
    def available_books(self):
        
        self.__available_books.clear()
        self.__not_available_books.clear()
        
        for i in range(len(self.lended_books)):
            self.__not_available_books.append(self.lended_books[i][1])
        
        for i in range(len(self.booksList)):
            if self.booksList[i] not in self.__not_available_books:
                self.__available_books.append(self.booksList[i])
                
        print("The available books : " + ', '.join(self.__available_books))        
            
        
    def book_lending(self,user_book):
        
        if user_book[1] in self.__available_books:
            self.lended_books.append(user_book)
            print(f'The book named ({user_book[1]}) is lended for {user_book[0]}')
            
        elif user_book[1] not in self.booksList:
            print("Sorry :( , We don't have this book yet")
        
        else:
            print('Sorry :( , The book is already lended')
            
     
    def add_book(self,book):
        
        self.booksList.append(book)
        print(f"The book named ({book}) has been added to bookslist")
        
        
    def return_borrowed_book(self,User_Book):
        
        if User_Book in self.lended_books:
            self.__available_books.append(User_Book[1])
            self.lended_books.remove(User_Book)
            print(f'{User_Book[0]} returned the book named ({User_Book[1]})')
            
        elif User_Book[1] not in self.__not_available_books:
            print(f"The book named ({User_Book[1]}) don't belong to this library")
            
        else:
            print("The name of the user is't correct")
            
        
    def show_lended_books(self):
        
        print("The lended books : ",end = '')
        for x in self.lended_books:
            print(x,end = ' ')
```
